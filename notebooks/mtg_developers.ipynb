```python
all_cards_path = os.path.join(os.getcwd(), 'data', 'all_cards.json')
all_cards_df = pd.DataFrame()
page_num = 0
more_pages = True
while more_pages:
    print('Getting page {}'.format(page_num))
    cards_url = 'https://api.magicthegathering.io/v1/cards?page={}'.format(page_num)
    res = requests.get(cards_url)
    cards_df = pd.DataFrame.from_dict(res.json()['cards'])
    print('cards_df shape: {}'.format(cards_df.shape))
    
    if cards_df.shape[0] > 0:
        all_cards_df = pd.concat([all_cards_df, cards_df], sort=True, ignore_index=True)
        print('all_cards_df shape: {}'.format(all_cards_df.shape))
        page_num += 1
    else:
        more_pages = False
        
all_cards_df.to_json(all_cards_path)
```
